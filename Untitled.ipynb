```python
ax = plt.axes(projection = "3d")
x = np.linspace(-100,100,3000)
y = np.linspace(-100,100,3000)
X , Y = np.meshgrid(x,y)

Z = 8 - X**2 - Y**2

ax.plot_surface(X,Y,Z , cmap = "plasma")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("x^2 + y^2")
plt.show()
```
